```python
house = house.drop('Unnamed: 0', axis=1)
man = man.drop('Unnamed: 0', axis=1)
woman = woman.drop('Unnamed: 0', axis=1)
tree = tree.drop('Unnamed: 0', axis=1)
```

```python
def find_boundbox(key):
    train_path = r"C:\Users\user\Desktop\23_kdt_hackathon\hackathon_data\266.AI 기반 아동 미술심리 진단을 위한 그림 데이터 구축\01.데이터\Training\02.라벨링데이터"
    dic = {'나무':'TL_나무', '남자':'TL_남자사람', '여자': 'TL_여자사람', '집': 'TL_집'}

    file_path = os.path.join(train_path, dic[key])
    files = os.listdir(file_path)

    for file_name in files:
        path = os.path.join(file_path, file_name)
        
        try:
            # JSON 파일을 열어서 내용을 읽고 파싱합니다.
            with open(path, "r", encoding="utf-8") as file:
                json_data = json.load(file)
                contents = json_data['annotations']['bbox']
                
                if key == '남자' or key == '여자':    
                    for content in contents:
                        if content['label'] == '팔':
                            arm_h.append(content['h'])
                            arm_w.append(content['w'])
                        elif content['label'] == '손':
                            hand_h.append(content['h'])
                            hand_w.append(content['w'])
                
                elif key =='나무':
                    for content in contents:
                        if content['label'] == '뿌리':
                            root_h.append(content['h'])
                            root_w.append(content['w'])
                
                elif key=='집':
                    for content in contents:
                        if content['label'] == '집전체':
                            house_h.append(content['h'])
                            house_w.append(content['w'])
        
        
        except Exception as e:
                print(f"파일 읽기 및 JSON 파싱 오류: {file_name}, 오류 메시지: {str(e)}")
                break
```

```python
arm_h = []
arm_w = []
hand_h = []
hand_w = []
root_h = []
root_w = []
house_h = []
house_w = []
find_boundbox('남자')
man['arm_h'] = pd.Series(arm_h)
man['arm_w'] = pd.Series(arm_w)
man['hand_h'] = pd.Series(hand_h)
man['hand_w'] = pd.Series(hand_w)

arm_h = []
arm_w = []
hand_h = []
hand_w = []
find_boundbox('여자')
woman['arm_h'] = pd.Series(arm_h)
woman['arm_w'] = pd.Series(arm_w)
woman['hand_h'] = pd.Series(hand_h)
woman['hand_w'] = pd.Series(hand_w)

find_boundbox('나무')
tree['root_h'] = pd.Series(root_h)
tree['root_w'] = pd.Series(root_w)

find_boundbox('집')
house['house_h'] = pd.Series(house_h)
house['house_w'] = pd.Series(house_w)
```

```python
# 사람: 팔, 손
# 나무: 뿌리
# 집: 집 전체
# 추가

tree.to_csv(r'C:\Users\user\Desktop\23_kdt_hackathon\hackathon_data\tree.csv', index=False)
man.to_csv(r'C:\Users\user\Desktop\23_kdt_hackathon\hackathon_data\man.csv', index=False)
woman.to_csv(r'C:\Users\user\Desktop\23_kdt_hackathon\hackathon_data\woman.csv', index=False)
house.to_csv(r'C:\Users\user\Desktop\23_kdt_hackathon\hackathon_data\house.csv', index=False)
```
